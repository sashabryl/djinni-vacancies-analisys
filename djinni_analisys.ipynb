```python
import datetime

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline

renaming_mapper = {"Machine_learning": "Machine Learning",
                   "Sql": "SQL",
                   "Fastapi": "FastAPI",
                   "Rest": "REST",
                   "Html": "HTML",
                   "Css": "CSS",
                   "Javascript": "JavaScript",
                   "Artificial_intelligence": "Artificial Intelligence"
                   }

timestamp = datetime.datetime.now().strftime("%d-%m-%Y, %H:%M:%S")
```

```python
def plot_csv(csv: str, title: str, filename: str = None):
    df = pd.read_csv(csv)
    df.columns = df.columns.str.capitalize()
    df.rename(columns=renaming_mapper, inplace=True)
    df.sum().sort_values().plot(kind="barh", figsize=(10, 8))
    plt.title(f"{title} at {timestamp}")
    plt.ylabel("Technology")
    plt.xlabel("Number of Mentions")
    if filename:
        plt.savefig(filename)
    plt.show()
```

```python
plot_csv(
    csv="general_vacancies.csv", 
    filename="general_vacancies_figure.png", 
    title="Most Desired Technologies from An Average PyDev"
)
```

```python
plot_csv(
    csv="junior_vacancies.csv", 
    filename="junior_vacancies_figure.png",
    title="Most Desired Technologies from A Junior PyDev"
)
```
